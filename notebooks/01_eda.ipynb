# Exploratory Data Analysis - Credit Card Fraud Detection

**Date**: 2025-01-10  
**Goal**: Understand the dataset characteristics, class distribution, and feature patterns

## Dataset
- **Source**: Kaggle Credit Card Fraud Detection  
- **Size**: 284,807 transactions  
- **Features**: 30 (Time, V1-V28, Amount)  
- **Target**: Class (0=Legitimate, 1=Fraud)

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings

warnings.filterwarnings('ignore')
sns.set_style('whitegrid')
plt.rcParams['figure.figsize'] = (12, 6)

%matplotlib inline
```

## Key Findings

After EDA, discovered:
1. Extreme class imbalance (0.17% fraud) - need SMOTE + F2 score
2. Amount has weak correlation with fraud - but log transform helps
3. Temporal patterns exist - extract Hour feature
4. V14, V12, V10 most predictive features
5. No missing values or data quality issues
